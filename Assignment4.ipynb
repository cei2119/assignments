```python
#question 1
def kelvinToCelsius(K):
    """This function takes in a kelvin temperature and output the equivalent temperature in celsius"""
    return K - 273.15
```

```python
#question 2
def tempToFahrenheit(temp, unit = 'celsius'):
    """This function takes in a temperature, either in celius or kelvin and converts it to fahrenheit
    If the temp is in kelvin it calls a previous function to convert it to celsius before converting to fahrenheit"""
    if unit == 'kelvin':
        cTemp = kelvinToCelsius(temp)
        fTemp = cTemp * 9/5 + 32
        return cTemp
    elif unit == 'celsius':
        fTemp = temp * 9/5 + 32
        return fTemp
    else:
        Print("Please enter temp, then specify either 'celsius' or 'kelvin'")
```

```python
#Question 3
print(tempToFahrenheit(0, "celsius"))
print(tempToFahrenheit(273.15, "kelvin"))
print(tempToFahrenheit(15))
```

```python
#Question 4
def fahrenheitToNewTemp(fTemp, type):
    """This function takes in a fahrenheit temperature and a type, celsius or kelvin,
    then returns the converted fahrenheit temp to either C or K depending on the user request"""
    cTemp = (fTemp - 32) * 5/9
    if type == 'celsius':
        return cTemp
    elif type == 'kelvin':
        return cTemp - 273.15
    else:
        print("User must enter celsius or kelvin as the type")
    
```

```python
print(fahrenheitToNewTemp(60, 'celsius'))
print(fahrenheitToNewTemp(60, 'kelvin'))
```

```python
fahrenheitToNewTemp(20, 'mars')
```

```python
#question 5
def imperialToMeters(feet, inches):
    """This function takes two arguments feet and inches and output height in terms of meters"""
    meters = feet * 0.3048 + inches * 0.0254
    return meters
```

```python
imperialToMeters(6, 3)
```

```python
#Question 6
def metersToImperial(meters):
    """This function takes in meters and outputs feet and remainder inches"""
    metersToInches = meters * 39.3701
    feet = int(metersToInches // 12)
    inches = int(metersToInches % 12)
    return (feet, inches)
```

```python
metersToImperial(1.905)
```

```python
#Question 7
#round trip conversion (p1 imperial to meters)
print(metersToImperial(3))
print(metersToImperial(1.9))
print(metersToImperial(2.76))
```

```python
#round trip conversian (p2 meters to imperial using the above values)
print(imperialToMeters(9, 10))
print(imperialToMeters(6, 2))
print(imperialToMeters(9, 0))
```
